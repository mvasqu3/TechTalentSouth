```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import pandas as pd
import numpy as np
data = pd.read_csv('company_sales_data.csv')
fig = plt.figure()
ax = plt.axes()
ax.plot(data)
plt.xlabel("Month")
plt.ylabel("Total Profit")
```

Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

```python
fig = plt.figure()
bx = plt.axes()
bx.plot(data.month_number,data.total_profit, ':r', marker='o',markerfacecolor='b', linewidth=3, label = 'Total Profit') 
#1 line stlye dotted and line color red
#6 and 7
plt.xlabel("Month Number") #3 xlabel
plt.ylabel("Total Profit") #4 ylabel
bx.legend(loc='lower right') #2 show legend in lower right
```

Exercise Question 3: Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
fig = plt.figure()
cx = plt.axes()
cx.plot(data.month_number,data.facecream, label='facecream')
cx.plot(data.month_number,data.facewash, label ='facewash')
cx.plot(data.month_number,data.toothpaste, label= 'toothpaste')
cx.plot(data.month_number,data.bathingsoap, label ='bathingsoap')
cx.plot(data.month_number,data.shampoo, label='shampoo')
cx.plot(data.month_number,data.moisturizer, label ='moisturizer')
cx.plot(data.month_number,data.total_units, label='total units')
plt.xlabel("Month Number") 
plt.ylabel("Unit Sold") 
cx.legend() 

```

Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

```python
fig = plt.figure()
dx = plt.axes()
dx.plot(data.month_number,data.toothpaste, 'o', label='Toothpaste' )
plt.xlabel("Month Number") 
plt.ylabel("Unit Sold") 
plt.title('Toothpaste Sales')
dx.legend() 
dx.grid(linestyle='-', color='r')
```
